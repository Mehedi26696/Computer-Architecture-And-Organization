## 1. Introduction
- The CPU does not work in isolation; it needs to communicate with **memory**, **input/output (I/O) devices**, and **peripherals**.
- This communication is essential for data exchange, instruction execution, and interaction with external components.

---

CPU communicates with external components via **buses**:
- **Data Bus**: Transfers data between CPU, memory, and I/O devices.
- **Address Bus**: Carries the address of memory or I/O locations to access.
- **Control Bus**: Carries control signals (e.g., Read/Write, Clock, Interrupts).

## 3. Types of External Communication

### 1. **Memory Communication**
- **Read/Write operations** to RAM or ROM.
- Uses MAR (Memory Address Register) and MDR (Memory Data Register).

### 2. **I/O Communication**
- Interaction with external devices like keyboard, mouse, display, etc.
- Managed using **I/O ports** or **memory-mapped I/O**.

### 3. **Interrupt Handling**
- External devices can send **interrupt signals** to request CPU attention.
- CPU stops current execution, saves state, and executes **Interrupt Service Routine (ISR)**.

---

## 4. Methods of Communication

### üîÅ Programmed I/O
- CPU actively waits and polls devices.
- Simple but inefficient (wastes CPU cycles).

### üîÅ Interrupt-Driven I/O
- Devices notify CPU when ready via interrupt.
- Efficient, as CPU can perform other tasks until needed.

### üîÅ Direct Memory Access (DMA)
- Allows devices to directly transfer data to/from memory without CPU intervention.
- CPU only sets up the transfer, freeing it for other tasks.

---

## 5. Communication Interfaces

| Interface       | Description |
|------------------|-------------|
| **PCI/PCIe**     | High-speed interface for attaching peripherals. |
| **USB**          | Universal interface for a wide range of devices. |
| **Serial/Parallel Ports** | Older forms of communication interfaces. |
| **Ethernet**     | Network communication for data transfer between systems. |

---

## 6. Synchronization and Handshaking

- Ensures both CPU and device are ready to communicate.
- **Handshaking signals** coordinate timing and status (e.g., Ready, Acknowledge).

---

## 7. Summary
- CPU communicates with memory and peripherals via buses and interfaces.
- Efficient communication methods like **DMA** and **interrupts** enhance performance.
- External communication is crucial for system functionality and responsiveness.
